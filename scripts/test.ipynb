```python
def preprocess_df(df: pd.DataFrame) -> pd.DataFrame:

    #cols_2_drop = ['Unnamed: 0','user_mentions','possibly_sensitive', 'original_text']
    df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date
    try:
        time = df[['timestamp','weekday','hour','minute','second']]
        flow = df[['flow1','flow2','flow3','totalflow']]
        ocup = df[['occupancy1','occupancy2','occupancy3']]
        #df = df.fillna(0)
    except KeyError as e:
        print("Error:", e)

    return time, flow, ocup
```
